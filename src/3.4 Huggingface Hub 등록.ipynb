# 3.4 Huggingface Hub 등록

### push_to_hub()

```python
import os
os.chdir('drive/MyDrive/Books')

import config as cfg
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained('outputs/trained_model')
model = AutoModelForSequenceClassification.from_pretrained('outputs/trained_model')

REPO_NAME = cfg.MY_REPO_NAME # ex) 'Seonghwan/my-bert-fine-tuned-psh1'
AUTH_TOKEN = cfg.MY_OWN_TOKEN
```

```python
model.push_to_hub(
    REPO_NAME,
    use_temp_dir=True,
    use_auth_token=AUTH_TOKEN
)
```

```python
tokenizer.push_to_hub(
    REPO_NAME,
    use_temp_dir=True,
    use_auth_token=AUTH_TOKEN
)
```
